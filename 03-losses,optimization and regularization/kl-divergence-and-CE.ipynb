How are they related?

Kl divergence is used to measure the distance between 2 probability distributions.
Cross entropy is also the same.

Let's try and understand with example.

Here is an example

If we have a die with 5 faces having
- `P(A) = 0.4`
- `P(B) = 0.2`
- `P(C) = 0.1`
- `P(D) = 0.1`
- `P(E) = 0.2`

and we draw a sequence
```
A A A A B B C D E E
```

The probability of getting this sequence is 
```
0.4 * 0.4 * 0.4 * 0.4 * 0.2 * 0.2 * 0.1 * 0.1 * 0.2 * 0.2 = 0.0000004096
```

we take the log sum to get a nicer number
```
= 4 * log(0.4) + 2 * log(0.2) + 2 * log(0.1) + 2 * log(0.2) 
   ---------------------------------------------------------
                         10
= -1.471
```


and take the negative of it
```
= 1.471
```

Here, `1.471` represents the entropy, which quantifies the uncertainty or the amount of information lost.



Now say we have to compare that distrbution against another distribution like this

```
P(A) = 0.4
P(B) = 0.2
P(C) = 0.1
P(D) = 0.1
P(E) = 0.2


Q(A) = 0.4
Q(B) = 0.1
Q(C) = 0.2
Q(D) = 0.2
Q(E) = 0.1
```

```
we now take cross entropy = -0.4 * log(0.4) + -0.2 * log(0.1) - 0.1 * log(0.2) - 0.1 * log(0.2) - 0.2 * log(0.1)
                          = 1.609
```

Now kl divergence is simply

```
cross entropy - entropy
= 1.609 - 1.471
= 0.138
```
